```python
size = (64, 64)

try:
    im = Image.open('bird.png')
    im.thumbnail(size)
    im.save('bird-thumb.png')
except IOError as e:
    print("cannot create thumbnail", e)
```

```python
im = Image.open("bird.png")
box = (100, 100, 400, 400)
region = im.crop(box)
region.save('cropped.png')
```

```python
region = region.transpose(Image.ROTATE_180)
im.paste(region, box)
im.save('transposed.png')
```

```python
from PIL import ImageFilter
im = Image.open("bird.png")
out = im.filter(ImageFilter.BLUR)
out.save('blured.png')
```

```python
r, g, b, a = im.split()
im1 = Image.merge("RGB", (b, g, r))
im1.save('withou-a.png')
```

```python
import numpy as np

def noise_filter(im):
    arr = np.array(im)
    noise = np.random.random(arr.shape)
    result = np.array(np.round(arr + noise), dtype=np.int8)
    return Image.fromarray(result, 'RGBA')
```

```python
im2 = noise_filter(im)
im2.save('with-noise.png')
```
